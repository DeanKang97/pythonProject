```python
import pandas as pd

true = pd.read_csv('/content/drive/MyDrive/종프/true_review_preprocessing.csv', dtype='string')
fake = pd.read_csv('/content/drive/MyDrive/종프/fake_review_preprocessing.csv', dtype='string')
```

```python
T = pd.DataFrame(true["블로그내용"])
T["label"] = 1

F = pd.DataFrame(fake["블로그내용"])
F["label"] = 0
```

```python
cv = CountVectorizer(tokenizer=get_noun)
tdm = cv.fit_transform(F["블로그내용"])

cv.vocabulary_
```

```python
cv = CountVectorizer(tokenizer=get_noun)
tdm = cv.fit_transform(T["블로그내용"])

cv.vocabulary_
```

```python
data = T.append(F)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.33, random_state=42)
```

```python
X_train = train["블로그내용"]
y_train = train["label"]

X_test = test["블로그내용"]
y_test = test["label"]
```

```python
from konlpy.tag import Twitter
from sklearn.feature_extraction.text import CountVectorizer

def get_noun(text):
    tokenizer = Twitter()
    nouns = tokenizer.nouns(text)
    return [n for n in nouns]
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline

text_clf_svm = Pipeline([('vect', CountVectorizer(tokenizer=okt.morphs)),
                         ('tfidf', TfidfTransformer(use_idf=True)), #TF-IDF를 사용해 피처를 만들 것인지 아니면 단어 빈도 자체를 사용할 것인지 여부
                         ('clf-svm', SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3, random_state=42))])

text_clf_svm = text_clf_svm.fit(X_train, y_train)
```

```python
predicted_svm = text_clf_svm.predict(X_test)
np.mean(predicted_svm == y_test)
```
